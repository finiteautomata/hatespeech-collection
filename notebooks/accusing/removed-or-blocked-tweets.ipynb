## Search for removed tweets

```python
%load_ext autoreload
%autoreload 2
from mongoengine import connect
from tweepyrate import create_apps
from hate_collector.models import Tweet

client = connect("hatespeech")
db = client["hatespeech"]
```

```python
errors_with_tweets = db.api_error.aggregate([ 
    { 
        "$lookup": {   
            "from": "tweet",   
            "localField": "tweet_id",   
            "foreignField": "_id", 
            "as": "tweet"
        }
    },
    {
        "$match": {
            "tweet": {"$not": {"$size": 0 } },
            "api_code": 422,
        }
    },
])

for i, error in enumerate(errors_with_tweets):
    print(f'{i} @{error["tweet"][0]["user"]["screen_name"]} :  ', error["tweet"][0]["text"])
```
