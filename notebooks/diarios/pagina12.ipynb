```python
%load_ext autoreload
%autoreload 2
import json
from mongoengine import connect
from tweepyrate import create_apps
from hatespeech_models import Tweet, Article

client = connect("hatespeech-news")
db = client["hatespeech-news"]


with open("../../dumps/news-argentina-v3.json") as f:
    arts = json.load(f)
```

```python
Article.objects.distinct('user')
```

```python
from tqdm.auto import tqdm
import datetime

newspapers = { "pagina12"}
newspaper_articles = []

for art in tqdm(arts):
    if art["user"] in newspapers:
        date = art["created_at"]["$date"]
        art["created_at"] = datetime.datetime.strptime(date, "%Y-%m-%dT%H:%M:%SZ")
        newspaper_articles.append(art)
    
len(newspaper_articles)
```

```python
import random

#newspaper_articles = sorted(newspaper_articles, key=lambda l: l["created_at"], reverse=True)
random.shuffle(newspaper_articles)


for art in newspaper_articles[:100]:
    print("="*80 +'\n\n')
    print(art["created_at"], '\n\n')
    print(art["title"], "\n\n")
    print(art["url"], "\n\n")
    print(":"*30, '\n\n')
    print(art["body"])
```
