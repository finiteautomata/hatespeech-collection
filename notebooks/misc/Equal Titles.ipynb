```python
%load_ext autoreload
%autoreload 2
from mongoengine import connect
from tweepyrate import create_apps
from hatespeech_models import Tweet, Article

client = connect("hatespeech-news")


print(f"Tenemos {Article.objects.count()} artículos")

users = Article.objects.distinct('user')

users
```

```python
arts = Article.objects(title="", user="LaVanguardia").order_by('created_at')

count = arts.count()

count
```

```python
from tqdm.auto import tqdm

for user in users:
    distinct_titles = Article.objects(user=user).distinct('title')
    
    print(f"{user:<15} --> {len(distinct_titles):<5} distintos, {Article.objects(user=user).count()} total")

    for title in tqdm(list(distinct_titles)):
        articles = Article.objects(title=title, user=user).order_by('created_at')
        count = articles.count()
        
        if count >= 20:
            print(f"{count:<2} - {title} (Too much!)")
            continue
        elif count >= 2:
                
            delta = articles[count-1].created_at - articles[0].created_at
            
            if delta.days > 2:
                total_comments = sum(len(a.comments) for a in articles)
                print(f"{count:<2} {delta} - {title} ({total_comments} comentarios)")
                print
```
