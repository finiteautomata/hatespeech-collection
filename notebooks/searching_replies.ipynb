## Buscar respuestas con seeds

Busquemos respuestas con seeds "racista", "mis√≥gino", etc

```python
tweets = []
replies = []
```

```python
import tweepy

class MyReplyListener(tweepy.StreamListener):
    def __init__(self, query):
        self._count = 0
        super().__init__()
        
    def on_status(self, status):
        global tweets
        try:
            tweets.append(status)
            if status.in_reply_to_status_id:
                replies.append(status)
            self._count += 1
            if self._count % 1000 == 0:
                print(f"{self._count / 1000:.2f}K tweets bajados")
                print(f"{self._count / 1000:.2f}K respuestas")
        except DuplicateKeyError as e:
            pass
    
    def on_error(self, status_code):
        print("Tenemos que esperar...")
        return True
```

```python
query = ["racista", "mis√≥gino", "machista"]
app = apps[0]

myStreamListener = MyReplyListener(query)
myStream = tweepy.Stream(auth = app.auth, listener=myStreamListener, tweet_mode= 'extended', language="es")
myStream.filter(track=query, is_async=True)
```
