```python
import time
from tweepyrate.streaming import create_queue, stream_query
import tweepy
from mongoengine import NotUniqueError

class MyTweetListener(tweepy.StreamListener):
    """
    Listener of Twitter Streaming API

    This class does not do any real time processing. It just pushes the
    received status to the respective queue, so another processes do the

    We do this to avoid being overwhelmed by the stream. See this link for further information:

    https://github.com/tweepy/tweepy/issues/448
    """
    def __init__(self, query, queue):
        self.query = query
        self.queue = queue
        super().__init__()

    def on_status(self, status):
        time.sleep(0.2)
        self.queue.put((status, self.query))

    def on_error(self, status_code):
        print(f"Error tipo: {status_code} para query {self.query}")
        print("Tenemos que esperar...")
        return True


class MyWorker:
    
    def __init__(self):
        self.count = 0
    def work(self, status, query):
        self.count += 1
        time.sleep(6)
        if self.count % 10 == 0:
            print(f"{self.count / 1000:.2f}K tweets")
            print(status.text)
            
queue = create_queue(1, MyWorker)
stream_query("bitch", apps[-1], queue, listener_class=MyTweetListener)
```
